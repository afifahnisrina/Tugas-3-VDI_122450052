```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Contoh data penjualan barang harian
data = {
    'Hari': ['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu'],
    'Penjualan (Unit)': [120, 150, 100, 180, 200, 220, 170]
}

df = pd.DataFrame(data)
```

```python
# Membuat bar chart sederhana dengan matplotlib
plt.figure(figsize=(8,5))
plt.bar(df['Hari'], df['Penjualan (Unit)'], color='skyblue')
plt.title('Penjualan Barang Harian')
plt.xlabel('Hari')
plt.ylabel('Penjualan (Unit)')
plt.show()
```

```python
# Membuat plot dengan seaborn
plt.figure(figsize=(8,5))
sns.lineplot(x='Hari', y='Penjualan (Unit)', data=df, marker='o')
plt.title('Penjualan Barang Harian')
plt.xlabel('Hari')
plt.ylabel('Penjualan (Unit)')
plt.grid(True)
plt.show()
```

```python
# Contoh data penjualan dua produk
data_heatmap = {
    'Hari': ['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu'],
    'Produk A': [120, 150, 100, 180, 200, 220, 170],
    'Produk B': [130, 140, 110, 190, 210, 230, 160]
}

df_heatmap = pd.DataFrame(data_heatmap)

# Membuat heatmap
plt.figure(figsize=(8,5))
sns.heatmap(df_heatmap.set_index('Hari'), annot=True, cmap='coolwarm')
plt.title('Penjualan Barang Harian per Produk')
plt.show()
```
